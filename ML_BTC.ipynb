```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
```

```python
df = pd.read_csv('/content/drive/MyDrive/BTC-USD.csv')
```

```python
df
```

```python
# Plotting a 2D line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], label='Closing Price')
plt.title('Bitcoin Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting a 3D scatter plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Open'], df['High'], df['Low'], c=df['Close'], cmap='viridis', s=20)
ax.set_xlabel('Open')
ax.set_ylabel('High')
ax.set_zlabel('Low')
ax.set_title('Bitcoin Price Characteristics')
plt.show()
```

```python
# Interactive 2D line chart using Plotly
fig = px.line(df, x='Date', y='Close', title='Bitcoin Closing Price Over Time (Interactive)',
              labels={'Close': 'Closing Price (USD)'}, line_shape='linear')
fig.update_xaxes(type='category', tickangle=45)
fig.show()

# Interactive 3D scatter plot using Plotly
fig = px.scatter_3d(df, x='Open', y='High', z='Low', color='Close', title='Bitcoin Price Characteristics (Interactive)',
                    labels={'Open': 'Open', 'High': 'High', 'Low': 'Low', 'Close': 'Closing Price (USD)'},
                    size_max=10, opacity=0.7)
fig.show()
```
